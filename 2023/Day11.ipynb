The researcher is trying to figure out the sum of the lengths of the shortest path between every pair of galaxies. However, there's a catch: the universe expanded in the time it took the light from those galaxies to reach the observatory.

In this example, after expanding the universe, the sum of the shortest path between all 36 pairs of galaxies is 374.

Expand the universe, then find the length of the shortest path between every pair of galaxies. What is the sum of these lengths?

````python
test_input = """...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#....."""

expanded_input = """....#........
.........#...
#............
.............
.............
........#....
.#...........
............#
.............
.............
.........#...
#....#......."""

result = 374
````

````python
from itertools import combinations


def parse_string(string: str) -> list[list[str]]:
    return [list(line) for line in string.splitlines()]


def get_all_galaxy_locations(space: list[list[str]]) -> list[tuple[int, int]]:
    return [
        (x, y)
        for x in range(len(space[0]))
        for y in range(len(space))
        if space[y][x] == "#"
    ]


def rows_to_be_expanded(space: list[list[str]]) -> list[int]:
    row_ids_to_expand = []
    for row_id, row in enumerate(space):
        if row.count(".") == len(row):
            row_ids_to_expand.append(row_id)
    return row_ids_to_expand


def columns_to_be_expanded(space: list[list[str]]) -> list[int]:
    column_ids_to_expand = []
    for column_id in range(len(space[0])):
        column = [row[column_id] for row in space]
        if column.count(".") == len(column):
            column_ids_to_expand.append(column_id)
    return column_ids_to_expand


def get_shortest_distances_in_expanded_space(
    space: list[list[str]], expand_by: int
) -> list[list[str]]:
    rows_to_expand = rows_to_be_expanded(space)
    columns_to_expand = columns_to_be_expanded(space)
    shortest_distances = []

    for galaxy1, galaxy2 in combinations(get_all_galaxy_locations(space), 2):
        rows_between = 0
        for row_id in range(min(galaxy1[1], galaxy2[1]), max(galaxy1[1], galaxy2[1])):
            if row_id in rows_to_expand:
                rows_between += 1
        columns_between = 0
        for column_id in range(
            min(galaxy1[0], galaxy2[0]), max(galaxy1[0], galaxy2[0])
        ):
            if column_id in columns_to_expand:
                columns_between += 1

        shortest_distances.append(
            abs(galaxy1[0] - galaxy2[0])
            + abs(galaxy1[1] - galaxy2[1])
            + rows_between * expand_by
            - rows_between
            + columns_between * expand_by
            - columns_between
        )

    return shortest_distances


print(sum(get_shortest_distances_in_expanded_space(parse_string(test_input), 2)) == 374)
````

````python
# part 1
with open("data/day11.txt") as f:
    input = f.read()

print(sum(get_shortest_distances_in_expanded_space(parse_string(input), 2)))
````
