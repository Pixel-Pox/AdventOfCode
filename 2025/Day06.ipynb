```python
with open("data/day06.txt") as f:
    lines = f.readlines()
    file = [line.split() for line in lines]

math_symbols = file[-1]
number_of_value_lines = len(file) - 1 # -1 for symbols line
```

```python
column_numbers = [numbers for numbers in zip(file[0], file[1], file[2], file[3], strict=True)]
column_numbers
```

```python
def cephalopod_math(symbol: str, numbers: list[int]) -> int:
    if symbol == '+':
        return sum(numbers)
    elif symbol == "*":
        result = 1
        for number in numbers:
            result *= number
        return result
        

assert len(column_numbers) == len(math_symbols)

total_part1 = 0
for number, symbol in zip(column_numbers, math_symbols, strict=True):
    total_part1 += cephalopod_math(symbol, [int(num) for num in number])

print(total_part1)
```

```python
from dataclasses import dataclass

with open("data/day06.txt") as f:
    lines = f.readlines()
    file = [list(line) for line in lines]

@dataclass
class VertNumbers:
    values: list[int]
    math_symbol: str = ""

    def get_total_value(self) -> int:
        if self.math_symbol == "+":
            return sum(self.values)
        elif self.math_symbol == "*":
            result = 1
            for value in self.values:
                result *= value
            return result
        else:
            raise ValueError(f"Unknown symbol {self.math_symbol}")


def is_split_col(col: tuple[str, ...]) -> bool:
    return all([c == " " for c in col])


all_math_problems: list[VertNumbers] = []

col_iter = iter(zip(*[file[i] for i in range(len(file))], strict=False))

v_n = VertNumbers(values=[])
while True:
    try:
        chars = next(col_iter)
        if is_split_col(chars):
            all_math_problems.append(v_n)
            v_n = VertNumbers(values=[])
            continue
        number = int("".join(chars[:4]).strip())
        symbol = v_n.math_symbol if chars[4] == " " else chars[4]
        v_n.values.append(number)
        v_n.math_symbol = symbol
    except StopIteration:
        all_math_problems.append(v_n)
        v_n = VertNumbers(values=[])
        break
```

```python
print(sum([v.get_total_value() for v in all_math_problems]))
```
