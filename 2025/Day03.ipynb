```python
with open("data/day03.txt") as f:
    file = f.read().split()

test_input = ["987654321111111", "811111111111119", "234234234234278", "818181911112111"]
```

```python
def get_join_of_two_largest_digits(digit_str: str) -> int:
    first_largest_idx = 0
    # we look for largest digit in all digits except for last one
    for idx, value in enumerate(digit_str[:-1]):
        if int(value) == 9:
            first_largest_idx = idx
            break
        if int(value) > int(digit_str[first_largest_idx]):
            first_largest_idx = idx

    second_largest_idx = first_largest_idx + 1
    for idx2, value2 in enumerate(digit_str[first_largest_idx + 1 :], first_largest_idx + 1):
        if int(value2) == 9:
            second_largest_idx = idx2
            break
        if int(value2) > int(digit_str[second_largest_idx]):
            second_largest_idx = idx2

    answer = int(digit_str[first_largest_idx] + digit_str[second_largest_idx])
    return answer


joltages: list[int] = []
for line in file:
    joltages.append(get_join_of_two_largest_digits(line))

print(f"Part 1 {sum(joltages)}")
```

```python
def iterate_for_max(string_subset: str) -> tuple[int, int]:  # returns index and value
    max_idx, max_value = 0, 0
    for idx, value in enumerate(string_subset):
        if int(value) == 9:
            return idx, value
        if int(value) > max_value:
            max_idx, max_value = idx, int(value)

    return max_idx, max_value


def get_join_of_two_largest_digits(digit_str: str, iterations: int) -> int:
    start_idx = 0
    values: list[str] = []
    for i in range(iterations):
        end_idx = -iterations + 1 + i
        string_subset = digit_str[start_idx:] if end_idx == 0 else digit_str[start_idx:end_idx]
        max_idx, result_value = iterate_for_max(string_subset)
        start_idx += max_idx + 1
        values.append(str(result_value))

    return int("".join(values))


joltages: list[int] = []
for line in file:
    joltages.append(get_join_of_two_largest_digits(line, 12))

print(f"Part 1 {sum(joltages)}")
```
