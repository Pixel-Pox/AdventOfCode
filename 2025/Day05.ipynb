```python
from dataclasses import dataclass

with open("data/day05.txt") as f:
    file = f.read().split()

ranges_str: list[str] = []
ids_to_check = []
for line in file:
    if "-" in line:
        ranges_str.append(line)
        continue
    elif len(line) == 0:
        continue

    ids_to_check.append(int(line))


@dataclass
class IDRange:
    min: int
    max: int

    def __hash__(self) -> int:
        return hash((self.min, self.max))
    
    def __lt__(self, other: "IDRange") -> bool:
        if self.min != other.min:
            return self.min < other.min
        return self.max < other.max


ranges: list[IDRange] = []

for range_str in ranges_str:
    ranges.append(IDRange(min=int(range_str.split("-")[0]), max=int(range_str.split("-")[1])))


test_ranges = [
    IDRange(min=3, max=5),
    IDRange(min=10, max=14),
    IDRange(min=16, max=20),
    IDRange(min=12, max=18)
]

test_ids_to_check = [1, 5, 8, 11, 17, 32]
print(len(ids_to_check))
```

```python
fresh_count = 0

for ingredient in ids_to_check:
    for id_range in ranges:
        if ingredient >= id_range.min and ingredient <= id_range.max:
            fresh_count += 1
            break

print(fresh_count)
```

```python
ranges = sorted(ranges)
```

```python
def combine_ranges(old_ranges: list[IDRange]) -> set[IDRange]:
    new_ranges: set = set()
    for rang in old_ranges:
        temp_min = 10000000000
        temp_max = 0
        for rang2 in old_ranges:
            if rang == rang2:
                continue
            if (
                (rang.min >= rang2.min and rang.max <= rang2.max)
                or (rang.min <= rang2.min and rang.max >= rang2.min)
                or (rang.max >= rang2.max and rang.min <= rang2.max)
            ):
                temp_min = min(rang.min, rang2.min)
                temp_max = max(rang.max, rang2.max)                

        if temp_max == 0: # No matches
            new_ranges.add(rang)
        else:
            new_ranges.add(IDRange(min=temp_min, max=temp_max))
    return new_ranges

prev_len = len(ranges)
ranges = combine_ranges(ranges)
while len(ranges) != prev_len:
    prev_len = len(ranges)
    ranges = combine_ranges(list(ranges))

total_number = 0
for r in ranges:
    total_number += r.max - r.min + 1

print(total_number)
```
