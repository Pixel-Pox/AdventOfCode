```python
test_file_ids = [
    ("11", "22"),
    ("95", "115"),
    ("998", "1012"),
    ("1188511880", "1188511890"),
    ("222220", "222224"),
    ("446443", "446449"),
    ("38593856", "38593862"),
    ("565653", "565659"),
    ("824824821", "824824827"),
    ("2121212118", "2121212124"),
]

invalid_ids = []

for min_range, max_range in file:
    for i in range(int(min_range), int(max_range) + 1):
        s_i = str(i)
        half_size = len(s_i) // 2
        for pattern_size in range(half_size, 0, -1):
            # If pattern size doesn't slice equally move next
            if len(s_i) % pattern_size != 0:
                continue
            # Split string into equal pieces based on pattern size
            splits = [s_i[i : i + pattern_size] for i in range(0, len(s_i), pattern_size)]
            if len(set(splits)) == 1:
                invalid_ids.append(i)
                break


print(sum(invalid_ids)) # 54446379122
```
