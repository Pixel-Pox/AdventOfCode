```python
with open("data/day04.txt") as f:
    _file = f.read().split()
    file = [list(string) for string in _file]

test_input = [
    "..@@.@@@@.",
    "@@@.@.@.@@",
    "@@@@@.@.@@",
    "@.@@@@..@.",
    "@@.@@@@.@@",
    ".@@@@@@@.@",
    ".@.@.@.@@@",
    "@.@@@.@@@@",
    ".@@@@@@@@.",
    "@.@.@@@.@.",
]
test_input = [list(string) for string in test_input]
```

```python
from dataclasses import dataclass


@dataclass
class Pos:
    x: int
    y: int


EMPTY = "."
PAPER = "@"

NEEDED_EMPTIES = 5

def check_8(arr2: list[list[str]], pos: Pos) -> bool:
    """Checks 8 adjecent positions to pos value.
    Returns number of them that are empty"""
    up_left = Pos(pos.x - 1, pos.y - 1)
    up = Pos(pos.x - 1, pos.y)
    up_right = Pos(pos.x - 1, pos.y + 1)
    left = Pos(pos.x, pos.y - 1)
    right = Pos(pos.x, pos.y + 1)
    down_left = Pos(pos.x + 1, pos.y - 1)
    down = Pos(pos.x + 1, pos.y)
    down_right = Pos(pos.x + 1, pos.y + 1)

    all_directions = [up_left, up, up_right, left, right, down_left, down, down_right]

    empties = 0
    for direction in all_directions: 
        # If out of bounds then empty
        if direction.x < 0 or direction.x == len(arr2):
            empties += 1
            continue
        if direction.y < 0 or direction.y == len(arr2[0]):
            empties += 1
            continue

        if arr2[direction.x][direction.y] == EMPTY:
            empties += 1
            continue

    return empties >= NEEDED_EMPTIES

accessable = 0
for idx_x in range(len(file)):
    for idx_y in range(len(file[0])):
        if file[idx_x][idx_y] == PAPER:
            accessable += check_8(file, Pos(x=idx_x, y=idx_y))

print(accessable)

```

```python
test_input = [
    "..@@.@@@@.",
    "@@@.@.@.@@",
    "@@@@@.@.@@",
    "@.@@@@..@.",
    "@@.@@@@.@@",
    ".@@@@@@@.@",
    ".@.@.@.@@@",
    "@.@@@.@@@@",
    ".@@@@@@@@.",
    "@.@.@@@.@.",
]
test_input = [list(string) for string in test_input]

total_removable = 0
while True:
    positions_to_remove: list[Pos] = []
    for idx_x in range(len(file)):
        for idx_y in range(len(file[0])):
            if file[idx_x][idx_y] == PAPER:
                accessable = check_8(file, Pos(x=idx_x, y=idx_y))
                if accessable:
                    positions_to_remove.append(Pos(idx_x, idx_y))
    
    if len(positions_to_remove) == 0:
        break

    total_removable += len(positions_to_remove)

    for position in positions_to_remove: 
        file[position.x][position.y] = EMPTY

print(total_removable)
```
